```python
# ---DRILL---
# Draw an 9x9 checkerboard using nested for loops
import drawsvg as draw
d = draw.Drawing(200, 200, origin = 'center')

c = 1
for i in range(0, 9):
    for j in range(0, 9):
        if c == 1:
            d.append(draw.Rectangle(-100 + (i * 20), -100 + (j * 20), 20, 20, fill = 'blue'))
            c = 0
        else:
            d.append(draw.Rectangle(-100 + (i * 20), -100 + (j * 20), 20, 20, fill = 'red'))
            c = 1

d
```
