```python
# --- DRILL ---
# Build a Course class
# 
# The constructor should take as input a number N
# and create a list of N Students. Each student
# can have a random string for a name, a random
# grade [0, 100] and a random height [150, 200]
# 
# The member function __str__ should print all of 
# the students in the class by calling the 
# Students classes __str__ function
# 
# Bonus: Plot a histogram of the grades for all N students
```

```python
import random
import string
import numpy as np
import matplotlib.pyplot as plt

# Student Class
class Student:
    def __init__(self, name, exam_grade, height_in_cm):
        self.name = name
        self.exam_grade = exam_grade
        self.height_in_cm = height_in_cm
    
    def __str__(self): # gaussian distribution more closely resembles classroom grade distribution
        return "(" + self.name + ", " + str(self.exam_grade) + ", " + str(self.height_in_cm) + ")"
    
    # add isFailing() function
    def isFailing(self):
        return self.getGrade() < 65
        '''
        # above code is the more concise version of this:
        if self.getGrade() <= 65:
            return True
        '''
    # accessor functions
    def getName(self):
        return self.name

    def getGrade(self):
        return self.exam_grade

    def getHeight(self):
        return self.height_in_cm
```

```python
# Course class
class Course:
    def __init__(self, N):
        self.N = N
        self.students = []
        for i in range(0, N):
            self.student = Student(''.join(random.choices(string.ascii_uppercase, k = 8)), random.gauss(75, 10), random.uniform(150, 200))
            self.students.append(self.student)
    
    def __str__(self):
        for i in range(0, len(self.students)):
            print(self.students[i])
        return "End of list"
    
    def histogram(self):
        grades = []
        for i in range(0, self.N):
            grades.append(self.students[i].getGrade())
        
        plt.hist(grades, 10)

# Driver code
math101 = Course(500)
print(math101)
math101.histogram()
```
