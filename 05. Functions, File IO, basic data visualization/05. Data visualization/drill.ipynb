```python
# ----- DRILL -----
# The csv file temperatures.csv consists of two columns [year, temp]
# corresponding to the change in global temperatures (C) relative to a baseline
# from 1880 to 2016. Read this csv file creating two lists consisting of the
# year and temperature. Plot the change in temperature as a function of year
# and label the axis accordingly. [NOTE: data will be read in as a string]
```

```python
# read csv file, make two lists consisting of year and temperature
import csv

temps = []
years = []
with open('temperatures.csv') as csvfile:
    csvdata = csv.reader(csvfile)
    next(csvdata, None) # skip the header
    for row in csvdata:
        years.append(float(row[0])) # year
        temps.append(float(row[1])) # temp
```

```python
# plot the change in temperature as a function of year and label the axis accordingly
import matplotlib.pyplot as plt

plt.plot(years, temps)
plt.xlabel('years')
plt.ylabel('temperature variation(C)')
plt.title('Change in global temperatures')
plt.show()
```
